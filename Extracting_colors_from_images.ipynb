```python
import cv2
import matplotlib.pyplot as plt
```

```python
def hist_rgb(rgb):
    return cv2.calcHist([rgb],[0],None,[256],[0,256])
```

```python
#RGB増指定でRGB値を返す関数
def show_rgb_value(path):

    img_data = cv2.imread(path)
    #RGB
    b_value= img_data[:,:,0]
    g_value= img_data[:,:,1]
    r_value= img_data[:,:,2]

    #ヒストグラム
    b_value_hist = hist_rgb(b_value)
    g_value_hist = hist_rgb(g_value)
    r_value_hist = hist_rgb(r_value)

    #図示
    plt.plot(r_value_hist, color='r', label="r")
    plt.plot(g_value_hist, color='g', label="g")
    plt.plot(b_value_hist, color='b', label="b")
    plt.legend()
    plt.show() 
    
    return b_value_hist, g_value_hist, r_value_hist
```

```python
r,g,b = show_rgb_value("/content/sample_data/sample.JPG")
```
